```python


import pandas as pd
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

# --- Define Variables and New Names ---
dependent_variable_old = 'Blood Pressure'
dependent_variable_new = 'Blood_Pressure' # Renamed for formula compatibility
independent_variable_old = 'Stress Level'
independent_variable_new = 'Stress_Level' # Renamed for formula compatibility

# Load the dataset
df = pd.read_csv("heart_disease.csv")

# 1. RENAME COLUMNS to remove spaces
df = df.rename(columns={
    dependent_variable_old: dependent_variable_new,
    independent_variable_old: independent_variable_new
})

# --- Data Preparation ---
# Drop rows with missing values
df_clean = df.dropna(subset=[dependent_variable_new, independent_variable_new]).copy()
# Ensure the independent variable is treated as categorical
df_clean[independent_variable_new] = df_clean[independent_variable_new].astype('category')

# --- Perform One-Way ANOVA ---

# Formula uses the safe, renamed column names
formula = f'{dependent_variable_new} ~ C({independent_variable_new})'

# Fit the OLS model
model = ols(formula, data=df_clean).fit()

# Calculate the ANOVA table
anova_table = anova_lm(model)

print("One-Way ANOVA Results:")
print(anova_table)
```
